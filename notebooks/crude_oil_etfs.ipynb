# Crude Oil Future


```python
from datetime import datetime
from pandas_datareader import DataReader
```

```python
end = datetime.now()
start = datetime(end.year - 1, end.month, end.day)
provider = "yahoo"
```

```python
symbols = [
    "USO", "UWT", "UCO", "DBO", "DWT",
    "OILU", "OIL", "SCO", "USL", "OILD",
    "DTO", "WTIU", "OILK", "UWTIF", "WTID",
    "DWTIF", "OILX", "OLEM"]

data_clf = DataReader("CL=F", provider, start, end)

g = globals()
for symbol in symbols:
    print(f"Fetching {symbol}...")
    g[f"data_{symbol.lower()}"] = DataReader(symbol, provider, start, end)
```

```python
data = data_clf[["Adj Close"]]
data = data.rename(columns={"Adj Close": "CL=F"})
for symbol in symbols:
    data_for_symbol = g[f"data_{symbol.lower()}"]
    data[symbol] = data_for_symbol["Adj Close"]
```
